```python
print(df3['PassOrFail'].value_counts())
df3['PassOrFail'] = df3['PassOrFail'].replace({"Y":0, "N":1})
df3['PassOrFail'].value_counts()
```

# Isolation Forest Modeling

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import MinMaxScaler
import numpy as np
from sklearn.metrics import confusion_matrix
```

```python
df_auto = df3.copy()
df_auto = df_auto.sort_values(by='TimeStamp')
df_auto.head()
```

```python
df_auto = df_auto.set_index('TimeStamp')
df_auto.head()
```

```python
train = df_auto[df_auto['PassOrFail'] == 0]
train = train.iloc[:-100]
test = pd.concat([df_auto[df_auto['PassOrFail'] == 1],train.iloc[-100:]])
#test = df_auto[df_auto['PassOrFail'] == 1]

print('train_value count: ')
print(train['PassOrFail'].value_counts())
print('=============================')
print('=============================')
print('=============================')
print('test_value_count: ')
print(test['PassOrFail'].value_counts())

X_train = train.drop('PassOrFail', axis=1)
X_test = test.drop('PassOrFail', axis=1)
print('=============================')
print(X_train.shape, X_test.shape)
```

```python
# 평가함수

def model_evaluation(label, predict):
    cf_matrix = confusion_matrix(label, predict)
    Accuracy = (cf_matrix[0][0] + cf_matrix[1][1]) / sum(sum(cf_matrix))
    Precision = cf_matrix[1][1] / (cf_matrix[1][1] + cf_matrix[0][1])
    Recall = cf_matrix[1][1] / (cf_matrix[1][1] + cf_matrix[1][0])
    Specificity = cf_matrix[0][0] / (cf_matrix[0][0] + cf_matrix[0][1])
    F1_Score = (2 * Recall * Precision) / (Recall + Precision)
    
    print("Accuracy: ", round(Accuracy,3))
    print("Precision: ", round(Precision,3))
    print("Recall: ", round(Recall,3))
    print("Specificity: ", round(Specificity,3))
    print("F1-Score: ", round(F1_Score,3))
```

```python
ss = StandardScaler()
ss.fit(X_train)
X_train = ss.transform(X_train)
X_test = ss.transform(X_test)
```

```python
import time

startTime = time.time()

if_clf = IsolationForest(contamination=0.3, max_features=1.0, 
                         max_samples=0.7, n_estimators=80, random_state=4)
if_clf.fit(X_train)

endTime = time.time() - startTime
print('훈련시간: ', round(endTime,2),'초')
print('==============================')

if_preds = if_clf.predict(X_test)
if_preds = np.array(pd.Series(if_preds).replace({1:0, -1:1}))
y_true = np.array((test['PassOrFail']))
print(confusion_matrix(y_true, if_preds))
print(model_evaluation(y_true, if_preds))
```
